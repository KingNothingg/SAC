### Уравнение поверхности второго порядка

```python
# Границы для трёхмерного графика
plot_range = 30.0
var("x y z")
f(x, y, z) = 7*x**2 + 8*x*y + 3*y**2 + 8*x*z + 6*y*z + 3*z**2 + 6*x + y + 7
show(u(x, y, z) == 0)
```

#### Построим исходную поверхность: 



```python
implicit_plot3d(f(x=x, y=y, z=z), (x, -plot_range, plot_range), (y, -plot_range, plot_range), (z, -plot_range, plot_range))
```

```python
# Из уравнения поверхности получим коэффициенты
a11 = 7
a22 = 3
a33 = 3

a12 = 8 / 2
a13 = 8 / 2
a23 = 6 / 2

a0 = 7
a1 = 6 / 2
a2 = 1 / 2
a3 = 0 / 2

A = matrix([
    [a11, a12, a13],
    [a12, a22, a23],
    [a13, a23, a33]
])

B = matrix([
    [a11, a12, a13, a1],
    [a12, a22, a23, a2],
    [a13, a23, a33, a3],
    [a1, a2, a3, a0]
])
show(A)
show(B)
```

```python
#ортогональные инварианты
t1 = A.trace()
t2 = A[0:2, 0:2].det() + A[[0, 2], [0, 2]].det() + A[1:3, 1:3].det()
d = A.det()
delta = B.det()
show("t1 = ", t1)
show("t2 = ", t2)
show("d = ", d)
show("delta = ", delta)
```

#### Исходя из вычесленных инвариантов, поверхность -- эллиптический парабалоид

```python
#найдем собственные значения
var("ev")
E = matrix([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])

eigen_values = []
for eigen_value in solve((A - ev * E).det() == 0, ev):
    print(n(eigen_value.rhs()))
    eigen_values.append(eigen_value.rhs())
```

#### Сравнение с eigenvalues: 

```python
for eigen_value in A.eigenvalues():
    print(eigen_value)
```

#### Значения совпали

```python
#коэффициенты для нового уравнения
a = n(1 / sqrt(-delta / (eigen_values[0]**2 * t2)))
b = n(1 / sqrt(-delta / (eigen_values[1]**2 * t2)))
```

```python
var("x1 y1 z1")
f_canonical(x1, y1, z1) = x1**2 * a + y1**2 * b - 2*z1
show(f_canonical(x1, y1, z1))
```

```python
implicit_plot3d(f_canonical(x1=x1, y1=y1, z1=z1), (x1, -plot_range, plot_range), (y1, -plot_range, plot_range), (z1, -plot_range, plot_range))
```
